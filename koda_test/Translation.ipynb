```python
from google.colab import drive
import os
import numpy as np
```

```python
# Connecting to Gdrive and read all filenames in subdirectory corpus
drive.mount('/content/gdrive/', force_remount=True)

# Read names of files in direcotry corpus
path = "/content/gdrive/MyDrive/NLPProject/corpus"
dir_list = os.listdir(path)
```

```python
# Then we read all *.txt files and extractall sentences from files
# Each sentence is in separate line, without any special tags
# For out project we use eccGigaFida corpus https://www.clarin.si/repository/xmlui/handle/11356/1035

allSentences = []

for i in dir_list:
    print(i)
    
    # samo preverim, če ima konico txt
    if not (i.endswith(".txt")):
        continue
        
    file = open(path+"/"+i, 'r', encoding="utf8")
    for j in file:
        
        # We extract just sencences, that are bigger than 10 characters.
        if len(j.strip()) > 10: 
            allSentences.append(j.strip().replace("'","").replace('"','').replace("“","").replace("»","").replace("«","").replace('”',''))
            
```

```python
# shranjevanje stavkov v izvorni obliki na disk
allSentences = np.asanyarray(allSentences,dtype=object)
np.save('/content/gdrive/MyDrive/NLPProject/data/untranslatedSentences',allSentences)
```

```python
import random
from google.colab import drive
import requests
from requests.structures import CaseInsensitiveDict
import googletrans
from googletrans import Translator
from json import JSONDecodeError
import time
import numpy as np
import deepl
```

```python
# Setting variables for translators

url = "http://localhost:4000/api/translate"
headers = CaseInsensitiveDict()
headers["Content-Type"] = "application/json"

translatorGoogle = Translator()

auth_key = "XXXX"  # Replace with your key
translatorDeepl = deepl.Translator(auth_key)
```

```python
# First we read untranslated and translated sentences
drive.mount('/content/gdrive/', force_remount=True)

untranslatedSentences = np.load('/content/gdrive/MyDrive/NLPProject/data/untranslatedSentences.npy',allow_pickle=True).tolist()
translatedSentences = []

# preberem še prevedene stavke - če obstaja
try:
    translatedSentences = np.load('/content/gdrive/MyDrive/NLPProject/data/translatedSentences.npy',allow_pickle=True).tolist()
except:
    print("There are no translated sentences yet")

print("We have",len(untranslatedSentences),"untranslated sentences, and",len(translatedSentences),"translated sentences")
```

```python
tags = dict({"original":0,"SLO_AN_CLARIN":1,"SLO_AN_GOOGLE":2,"SLO_AN_DEEPL":3,
             "SLO_AN_SLO_CLARIN_CLARIN":4, "SLO_AN_SLO_GOOGLE_GOOGLE":5, "SLO_AN_SLO_DEEPL_DEEPL":6})
```

```python
for i in range(10):
    
    print()
    print(i)
    
    indeks = random.randrange(len(untranslatedSentences))
    sentenceForTranslation = untranslatedSentences[indeks]
    
    print(sentenceForTranslation)
    
    # If sentence is too short or too big, then we oimt it
    if sentenceForTranslation.count(' ') > 20 or sentenceForTranslation.count(' ') < 5:
        print("Too short or too long sentence.")
        continue
    
    # Variables for translated sentences
    sentenceSloAnClarin = ""
    sentenceSloAnSloClarinClarin = ""
    sentenceSloAnGoole = ""
    sentenceSloAnSloGoogleGoogle = ""
    sentenceSloAnDeepl = ""
    sentenceSloAnSloDeeplDeepl = ""

    # New line for table of translated sentences
    lineForTranslatedSentencesTable = [""]*7
    lineForTranslatedSentencesTable[0] = sentenceForTranslation

    result="test"
    
    """
    # Slovene - English - Slovene translattion with Clarin
    try:
        data = '{ "src_language":"sl","tgt_language":"en","text":"'+sentenceForTranslation+'"}'
        data = data.encode("utf-8")
        resp = requests.post(url, headers=headers, data=data)
        sentenceSloAnClarin = eval(resp.text)['result']
        lineForTranslatedSentencesTable[tags["SLO_AN_CLARIN"]] = sentenceSloAnClarin
        data = '{ "src_language":"en","tgt_language":"sl","text":"'+sentenceSloAnClarin+'"}'
        data = data.encode("utf-8")
        resp = requests.post(url, headers=headers, data=data)
        sentenceSloAnSloClarinClarin = eval(resp.text)['result']
        lineForTranslatedSentencesTable[tags["SLO_AN_SLO_CLARIN_CLARIN"]] = sentenceSloAnSloClarinClarin
    except:
        print("Error CLARIN")
    """
        
    
    # Slovene - English - Slovene translattion with Google Translate
    try:
        result = translatorGoogle.translate(sentenceForTranslation, src='sl', dest='en')
        sentenceSloAnGoole = result.text
        lineForTranslatedSentencesTable[tags["SLO_AN_GOOGLE"]] = sentenceSloAnGoole
        result = translatorGoogle.translate(sentenceSloAnGoole, src='en', dest='sl')
        sentenceSloAnSloGoogleGoogle = result.text
        lineForTranslatedSentencesTable[tags["SLO_AN_SLO_GOOGLE_GOOGLE"]] = sentenceSloAnSloGoogleGoogle
    except:
        print("Error GOOGLE")

        
    # Slovene - English - Slovene translattion with Deepl
    try:
        result = translatorDeepl.translate_text(sentenceForTranslation, source_lang="SL", target_lang="EN-GB")
        sentenceSloAnDeepl = result.text
        lineForTranslatedSentencesTable[tags["SLO_AN_DEEPL"]] = sentenceSloAnDeepl
        result = translatorDeepl.translate_text(sentenceSloAnDeepl, source_lang="EN", target_lang="SL")
        sentenceSloAnSloDeeplDeepl = result.text
        lineForTranslatedSentencesTable[tags["SLO_AN_SLO_DEEPL_DEEPL"]] = sentenceSloAnSloDeeplDeepl
    except:
        print("Error DEEPL")

    untranslatedSentences.pop(indeks)
    translatedSentences.append(lineForTranslatedSentencesTable)

    time.sleep(5)
 
```

```python
np.save('/content/gdrive/MyDrive/NLPProject/data/untranslatedSentences',np.asanyarray(untranslatedSentences,dtype=object))
np.save('/content/gdrive/MyDrive/NLPProject/data/translatedSentences',np.asanyarray(translatedSentences,dtype=object))
```
